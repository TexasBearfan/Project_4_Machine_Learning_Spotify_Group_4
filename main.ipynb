```python
df=pd.read_csv(file,delimiter=';')
df.head()
```

```python
df=df[['Rank','Date','Artists','Nationality','Song URL','Continent','Artist (Ind.)','Title','Danceability','Energy','Loudness','Speechiness','Acousticness','Instrumentalness','Valence','id']]
df.head()
```

```python
df['Dated_date']=pd.to_datetime(df['Date'])
df.head()
```

```python
dated_Df=df.loc[df['Dated_date']>='2022-05-30']
dated_Df.tail()
```

```python
dated_Df.info()
```

```python
dated_Df['Dated_date'].nunique()
```

```python
def date2num(date_time):
  d, m, y = date_time.split('/')
  num = int(d)*10 + int(m)*100
  return num
```

```python
import numpy as np
date_features = []
for d in list(dated_Df['Date']):
  date_features.append(date2num(d))
date_features = np.array(date_features)
date_features_normalized = (date_features - np.min(date_features))/(np.max(date_features) - np.min(date_features))
```

```python
dated_Df['Date_value'].nunique()
```

```python
dated_Df['Date_value']=date_features_normalized
dated_Df.head()
```

```python
grouped_df=dated_Df.groupby('Date_value')['Danceability'].mean()
grouped_df.head()
```

```python
grouped_df.plot.line(x='Date_value',y='Danceability')
```

```python
dated_Df.to_csv('Big_Data_2022_05_30.csv')
```
