```python
import pandas as pd
```

```python
file='Spotify_Dataset_V3.csv'
```

```python
df=pd.read_csv(file,delimiter=';')
df.head()
```

```python
df=df[['Rank','Date','Title','Danceability','Energy','Loudness','Speechiness','Acousticness','Instrumentalness','Valence','id']]
df.head()
```

```python
df=df.drop_duplicates()
df.head()
```

```python
df['Date'].nunique()
```

```python
df['Date'].unique()
```

```python
df['Dated_date']=pd.to_datetime(df['Date'])
df.head()
```

```python
#between may 30 2022 to may 29 2023
dated_Df=df.loc[df['Dated_date']>='2022-05-30']
dated_Df.tail()
```
